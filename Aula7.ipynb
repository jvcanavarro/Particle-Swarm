```python
# Random Walk/PSO Para resolver S. de Equações c/ 3 variáveis.

'''Fontes
https://jamesmccaffrey.wordpress.com/2015/06/09/particle-swarm-optimization-using-python/
https://nathanrooy.github.io/posts/2016-08-17/simple-particle-swarm-optimization-with-python/
Particle swarm algorithm for solving systems of nonlinear equations - MajidJaberipour
'''

from __future__ import division
import random as rand
import numpy as np
import matplotlib.pyplot as plt
import math

def med_sistema(x):
	x1,x2,x3 = x[0],x[1],x[2]

	#Equações.
	eqs = []
	eqs.append(3*x1 + 2*x2 + 4*x3) # = 1
	eqs.append(x1 + x2 + 2*x3) # = 2
	eqs.append(4*x1 + 3*x2 - 2*x3) # = 3

	#Erro relativo.
	med = []
	for i in range (3):
		j = (i + 1 - eqs[i])/(i + 1)
		med.append(abs(j))

	return np.mean(med), x

def createSequence(TAM):
	i = 0
	medEquation = []
	while i < TAM:
		sequence = [round(rand.uniform(-10.,10.),5)for l in range (3)]
		medEquation.append(med_sistema(sequence))
		i+=1
	return medEquation

def findBestCase(medEquation,numIterations,bestCase):

	# Condições de Parada.
	if 0.95 < bestCase[0] <= 1:
		print("Condição de Parada I : Resultado dentro do intervalo esperado encontrado.")
		return bestCase
	elif numIterations <= 0:
		print("Condição de Parada II : Número de iterações chegou ao limite.")
		return bestCase

	# Analisa as médias.
	for i in range (50):
		aux = medEquation[i]
		if 1-abs(aux[0]) > 1-abs(bestCase[0]) and 0.95 < aux[0] <= 1:
			bestCase = aux
	
	# Retira os piores casos.
	medEquation.sort()
	medEquation = medEquation[:len(medEquation)-25]
	medEquation += createSequence(25)

	return findBestCase(medEquation,numIterations-2,bestCase)

numExec = 5
i = 0
j = 0
melhores = []
for i in range(numExec):
	numIterations = 1000
	medEquation = createSequence(50)	
	medEquation.sort()
	bestCase = medEquation[0]
	melhores.append(findBestCase(medEquation,numIterations, bestCase))
	perc = melhores[i][0] * 100
	print("Execução %d: %f (~%.1f %%) de proximidade na sequência %s.\n" %(i + 1 ,melhores[i][0],round (perc,1),melhores[i][1]))

# Média e Desvio Padrão das 5 Execuções.
medias = [x[0] for x in melhores]
print("Média (5 Execuções) : %f (~%.1f%%)" %(np.mean(medias)*100,round(np.mean(medias)*100,1)))
print("Desvio Padrão :%.4f%%" %(np.std(medias)*100))

# Resultado real.
x = np.linalg.solve([ [3, 2, 4], [1, 1, 2], [4, 3, -2] ],[ 1, 2, 3 ])

# Calcula a proximidade entre as Execuções e o.Resultado Real (x).
solucoes = [x[1] for x in melhores]
media_geral = []
for i in range (5):
	media_unica = []
	for j in range (3):
		prox = (solucoes[i][j])-x[j]/solucoes[i][j]
		media_unica.append(abs(prox))
	media_geral.append(np.mean(media_unica))

print("\nAproximação das 5 execuções com o resultado exato X(%s): %.2f%%" %(x,np.mean(media_geral)))


# Plotar os 5 BestCases/Resultado Exato.

y = np.arange(5)
plt.subplot(2,1,1)
plt.bar(y,medias,color='c')
plt.title("RW")
plt.show()

#plt.savefig("5melhores.png")


'''PSO'''
def func1(x):
	x1,x2,x3= x[0],x[1],x[2]

	eqs = []
	eqs.append(3*x1 + 2*x2 + 4*x3) # = 1
	eqs.append(x1 + x2 + 2*x3) # = 2
	eqs.append(4*x1 + 3*x2 - 2*x3) # = 3

	med = []
	for i in range (3):
		j = (i + 1 - eqs[i])/(i + 1)
		med.append(abs(j))

	return np.mean(med)

#--- MAIN ---------------------------------------------------------------------+

class Particle:
    def __init__(self,x0):
        self.position_i=[]          # particle position
        self.velocity_i=[]          # particle velocity
        self.pos_best_i=[]          # best position individual
        self.err_best_i=-1          # best error individual
        self.err_i=-1               # error individual

        for i in range(0,num_dimensions):
            self.velocity_i.append(rand.uniform(-1,1))
            self.position_i.append(x0[i])

    # evaluate current fitness
    def evaluate(self,costFunc):
        self.err_i=costFunc(self.position_i)

        # check to see if the current position is an individual best
        if self.err_i<self.err_best_i or self.err_best_i==-1:
            self.pos_best_i=self.position_i
            self.err_best_i=self.err_i
                    
    # update new particle velocity
    def update_velocity(self,pos_best_g):
        w=0.5       # constant inertia weight (how much to weigh the previous velocity)
        c1=1        # cognative constant
        c2=2        # social constant
        
        for i in range(0,num_dimensions):
            r1=rand.random()
            r2=rand.random()
            
            vel_cognitive=c1*r1*(self.pos_best_i[i]-self.position_i[i])
            vel_social=c2*r2*(pos_best_g[i]-self.position_i[i])
            self.velocity_i[i]=w*self.velocity_i[i]+vel_cognitive+vel_social

    # update the particle position based off new velocity updates
    def update_position(self,bounds):
        for i in range(0,num_dimensions):
            self.position_i[i]=self.position_i[i]+self.velocity_i[i]
            
            # adjust maximum position if necessary
            if self.position_i[i]>bounds[i][1]:
                self.position_i[i]=bounds[i][1]

            # adjust minimum position if neseccary
            if self.position_i[i]<bounds[i][0]:
                self.position_i[i]=bounds[i][0]
        
class PSO():
  def __init__(self,costFunc,x0,bounds,num_particles,maxiter,iteration):
      global num_dimensions
      num_dimensions=len(x0)
      err_best_g=-1                   # best error for group
      pos_best_g=[]                   # best position for group

      # establish the swarm
      swarm=[]
      for i in range(0,num_particles):
          swarm.append(Particle(x0))

      # begin optimization loop
      i=0
      while i<maxiter:
          
          # cycle through particles in swarm and evaluate fitness
          for j in range(0,num_particles):
              swarm[j].evaluate(costFunc)

              # determine if current particle is the best (globally)
              if swarm[j].err_i<err_best_g or err_best_g==-1:
                  pos_best_g=list(swarm[j].position_i)
                  err_best_g=float(swarm[j].err_i)
          
          # cycle through swarm and update velocities and position
          for j in range(0,num_particles):
              swarm[j].update_velocity(pos_best_g)
              swarm[j].update_position(bounds)
          i+=1
      print ('\nResultado Final (PSO) - Nº de Partículas:%d ; Iterações Máximas : %d'%(num_particles,maxiter))
      print ('Melhor Sequência (Exec. %d): %s'%(iteration,pos_best_g))
      print ('Erro :%f (~%.2f%%)'%(err_best_g,round(err_best_g,2)*100))
      erro = 1 - abs(err_best_g)
      global medias
      medias.append(erro)
      
pso_Exec = 5
num_particles = 500
maxiter = 1000
bounds=[(-10.,10.),(-10.,10.),(-10.,10.)]

for i in range(pso_Exec):
  #initial = [5,5,5]
  initial = [round(rand.uniform(-10,10),4) for i in range (3)]
  PSO(func1,initial,bounds,num_particles,maxiter,i+1)

# Plotar PSO

z = np.arange(len(medias))

plt.subplot(2 ,1 ,2)
plt.bar( z ,medias ,color='c' )
plt.title(" RW x PSO ")
plt.tight_layout()
plt.grid( linestyle = ':' , color = 'k')
plt.show()

#plt.savefig("5melhores_PSO.png")
```
